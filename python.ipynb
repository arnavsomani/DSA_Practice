```python
# 1.
# Given an integer array nums, return true if any value appears at least twice 
# in the array, and return false if every element is distinct.

# https://leetcode.com/problems/contains-duplicate/description/

nums = [ 1, 2, 3, 4, 5, 6 ]

class Solution():
    def __init__(self, nums: list):
        self.nums = nums

    def contains_duplicate(self) -> bool:
        l = []
        for i in range(len(self.nums)):
            if self.nums[i] not in l:
                l.append(self.nums[i])

        return bool(len(l)==len(self.nums))
    
sol = Solution(nums)
sol.contains_duplicate()
```

```python
#2. 
# Given an array of integers nums and an integer target, return 
# indices of the two numbers such that they add up to target.

# You may assume that each input would have exactly one solution, 
# and you may not use the same element twice.

# You can return the answer in any order.

# Example 1:

# Input: nums = [2,7,11,15], target = 9
# Output: [0,1]
# Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

# https://leetcode.com/problems/two-sum/

nums = [2,7,11,15]
nums = [3,2,4]
target = 9
target = 6 

class Solution:
    def __init__(self, nums, target):
        self.nums = nums
        self.target = target

    def two_sum(self):
        seen = {}  # number -> index

        for i, num in enumerate(self.nums):
            needed = self.target - num

            if needed in seen:
                return [seen[needed], i]

            seen[num] = i


sol = Solution(nums,target)
sol.two_sum()
```

```python
# 3. 
# https://leetcode.com/problems/valid-anagram/description/

# Given two strings s and t, return true if t is an anagram of s, and false otherwise.
# Example 1:
# Input: s = "anagram", t = "nagaram"
# Output: true

# Example 2:
# Input: s = "rat", t = "car"
# Output: false

s = "anagram"
t = "nagaram"

class Solution:
    def __init__(self, s, t):
        self.s = s
        self.t = t
    
    def two_string(self):
        total_s = {}
        total_t = {}

        for i, num in enumerate(self.s):
            if num not in total_s:
                total_s[num] = 1
            else:
                total_s[num]+= 1
        
        for i, num in enumerate(self.t):
            if num not in total_t:
                total_t[num] = 1
            else:
                total_t[num]+= 1

        return bool(total_s == total_t)

sol = Solution(s,t)
sol.two_string()
```

```python
# 4

# https://leetcode.com/problems/group-anagrams/description/

# Given an array of strings strs, group the anagrams together. 
# You can return the answer in any order. 

# Example 1:
# Input: strs = ["eat","tea","tan","ate","nat","bat"]
# Output: [["bat"],["nat","tan"],["ate","eat","tea"]]

# Explanation:
# There is no string in strs that can be rearranged to form "bat".
# The strings "nat" and "tan" are anagrams as they can be rearranged to form each other.
# The strings "ate", "eat", and "tea" are anagrams as they can be rearranged to form each other.

# Example 2:
# Input: strs = [""]
# Output: [[""]]

# Example 3:
# Input: strs = ["a"]
# Output: [["a"]]

strs = ["eat","tea","tan","ate","nat","bat"]

class Solution:
    def __init__(self, strs):
        self.strs = strs

    def group_anagrams(self):
        
        hash_map = {}

        for i, num in enumerate(self.strs):
            sorted_word = "".join (sorted(num)) 

            if sorted_word not in hash_map:
                hash_map[sorted_word] = []
            
            hash_map[sorted_word] += [num]

        return [sorted(i) for i in hash_map.values()][::-1]
            
sol = Solution(strs)
sol.group_anagrams()
```

```python
# 5

# https://leetcode.com/problems/top-k-frequent-elements/

# Given an integer array nums and an integer k, return the k most frequent elements. 
# You may return the answer in any order.

# Example 1:
nums = [1,1,1,2,2,3]
k = 2
# Output: [1,2]

# Example 2:
# Input: nums = [1], k = 1
# Output: [1]

# Example 3:
# Input: 
nums = [1,2,1,2,1,2,3,1,3,2]
k = 2
# Output: [1,2]

class Solution:
    def __init__(self, nums, k):
        self.nums = nums 
        self.k = k
    
    def top_freq(self):

        hash_map = {}

        for i, num in enumerate(self.nums):
            if num not in hash_map:
                hash_map[num] = 1
            else:
                hash_map[num] += 1
        
        sorted_items = sorted(hash_map.items(), key = lambda x : x[1], reverse=True)

        return [ i for i,k in sorted_items[:self.k]]

sol = Solution(nums, k)
sol.top_freq()
```
